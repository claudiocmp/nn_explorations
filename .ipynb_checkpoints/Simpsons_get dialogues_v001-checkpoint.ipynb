```python
import scrapy
from scrapy.crawler import CrawlerProcess
from scrapy.utils.project import get_project_settings
from scrapy.http import TextResponse
import requests
import os
from os.path import join
```

```python
process = CrawlerProcess(get_project_settings())
```

Create Spider class

```python
class SimpsonsSpider(scrapy.Spider):
    name = 'SimpsonsSpider'
    
    def __init__(self):
        self.url = "https://www.springfieldspringfield.co.uk/view_episode_scripts.php?tv-show=the-simpsons&episode=" #s01e01    
    
    def start_requests(self):
        
        urls = []

        for s in range(1, 29+1):
            for e in range(1, 25+1):
                urls.append("{0}s{1:02d}e{2:02d}".format(self.url, s, e))
        
        for u in urls:
            yield scrapy.Request(url=u, callback=self.parse)

    def parse(self, response):
        
        page = response.url.split("&episode=")[-1]
        
        filename = '%s.txt' % page
        
        #for resp in response.xpath('//*[@id="content_container"]/div[2]/div[2]/div[2]/div[1]'):
        txt = response.xpath('//*[@id="content_container"]/div[2]/div[2]/div[2]/div[1]').extract_first()
        txt = txt.split('<div class="scrolling-script-container">')[1]
        txt = txt.replace('</div>', '')
        txt = txt.replace('<br>', '\r\n')

        with open(join('txts', filename), 'x') as f:
            f.write(txt)
            
        self.log('Saved file %s' % filename)

        #yield txt
            
            #yield {
             #   'text': res.extract_first(),
#                'author': quote.css('small.author::text').extract_first(),
#                'tags': quote.css('div.tags a.tag::text').extract(),
            #}
        
        #with open(filename, 'wb') as f:
         #   f.write(response.body)
        #self.log('Saved file %s' % filename)
```

Start the crawler

```python
process.crawl(SimpsonsSpider)
process.start() # the script will block here until the crawling is finished
```

Compound the dialogues from all episodes into one

```python
with open(join('all.txt'), 'wb') as f:
    for file in os.listdir('txts'):
        with open(join('txts', file), 'r') as i:
            f.write(i.buffer.read())
```
